## **Example 1: List Manipulation**

```python
# --- 1. Defining a List ---
# Creating a list with mixed data types
tools = ["Laptop", "Monitor", "Keyboard", "Mouse"]

```

```python
# --- 2. Indexing and Slicing ---
# Accessing elements by position (0-based)
print(tools[0])     # Output: Laptop
print(tools[-1])    # Output: Mouse (Last item)

# Slicing: [Start : Stop] - Note: Stop index is exclusive
sub_list = tools[1:3]
print(sub_list)     # Output: ['Monitor', 'Keyboard']
```

```python


# --- 3. Modifying Elements ---
# Changing a value at a specific index
tools[2] = "Webcam"
print(tools)        # Output: ['Laptop', 'Monitor', 'Webcam', 'Mouse']

```

```python
# --- 4. Adding Elements ---
# Use append() to add to the end of the list
tools.append("Desk")

# Use insert() to add at a specific index
tools.insert(1, "Headphones")
print(tools)        # Output: ['Laptop', 'Headphones', 'Monitor', 'Webcam', 'Mouse', 'Desk']

```

```python
# --- 5. Removing Elements ---
# Use remove() to delete a specific value by name
tools.remove("Monitor")
print(tools)        # Output: ['Laptop', 'Headphones', 'Webcam', 'Mouse', 'Desk']

```

```python
# --- 6. Checking Length ---
# Using the len() function to see total count
total_items = len(tools)
print(total_items)  # Output: 5
```

### **Copying Lists vs Creating References in Python**

### **Creating a Reference (Not a Copy)**

Assigning one list variable to another **does not create a new list**.  
Instead, it creates **another reference to the same list object in memory**.

When two variables reference the same list, modifying one variable will also modify the other because they point to the same object.

```python
# This does NOT copy a list

original_list = [1, 2, 3]

# new_reference points to the same list object
new_reference = original_list

# Modify the list through the new reference
new_reference.append(4)

print("Original List:", original_list)
print("New Reference:", new_reference)
print(original_list == new_reference)
```

### Key Point

- `new_reference = original_list` **does not copy the list**.
- Both variables point to **the same object in memory**.
- Changing one variable **changes the other**.

### **Properly Copying a List**

To create a **separate list object with the same contents**, we can use the `.copy()` method.

A copied list is independent from the original list, so modifying the copied list will not affect the original.

```python
original_list = ["A", "B", "C"]

# Create a true copy of the list
copied_list = original_list.copy()

# Modify the copied list
copied_list.append(4)

print("Original List:", original_list)
print("Copied List:", copied_list)
print(original_list == copied_list)
```

### Key Point

- `.copy()` creates a **new list object**.
- The copied list contains the same elements but is stored in a **different memory location**.
- Changes made to the copied list **do not affect the original list**.

### **Variables reference the same list,**

If two variables reference the same list, modifying one will also modify the other.

```python
furniture_list1 = ["table", "chair", "lamp"]

# This creates another reference, not a copy
furniture_list_append = furniture_list1

furniture_list_append.append("armoire")

print("---------- furniture----------")
print(furniture_list1)
print("---------- furniture_list_append----------")
print(furniture_list_append)
```

## **Example 2: Using .reverse() and .extend() methods**

## **Python List Methods Overview: `.reverse()` and `.extend()`**

### **`.reverse()`**

**Purpose:**  
The **`.reverse()`** method reverses the order of the elements in a list.

**Key Points**
- The operation is performed **in place**, meaning the original list is modified.
- It **does not create a new list**.
- The method **returns `None`** because it directly updates the existing list.
- Commonly used when the order of elements needs to be flipped without creating additional memory overhead.

---

### **`.extend()`**

**Purpose:**  
The **`.extend()`** method adds elements from another iterable to the end of a list.

**Key Points**
- The method **modifies the original list**.
- It accepts **any iterable object**, such as lists, tuples, sets, or strings.
- Each element from the iterable is **added individually** to the list.
- The method **returns `None`** since the list is updated in place.

---

### **Summary**

| Method | Function | Behavior |
|------|------|------|
| `.reverse()` | Reverses the order of elements in a list | Modifies the original list in place |
| `.extend()` | Adds elements from another iterable to a list | Expands the list by inserting each element individually |


This following function accepts two variables, each containing a list of years. A current **"recent_first"** list contains [2022, 2018, 2011, 2006].
An older **"recent_last"** list contains [1989, 1992, 1997, 2001].
The lists need to be combined with the years in chronological order.

- Reverse the order of a list using the .reverse() method.
- Combine two lists using the .extend() method.

```python
def record_profit_years(recent_first, recent_last):
    # Reverse the order of the "recent_first" list so that it is in chronological order.
    recent_first.reverse()

    # Extend the "recent_last" list by appending the newly reversed "recent_first" list.
    recent_last.extend(recent_first)

    # Return the "recent_last", which now contains the two lists combined in chronological order.
    return recent_last
```

```python
# Assign the two lists to the two variables to be passed to the record_profit_years() function.
recent_first = [2022, 2018, 2011, 2006]
recent_last = [1989, 1992, 1997, 2001]

# Call the record_profit_years() function and pass the two lists as  parameters.
print(record_profit_years(recent_first, recent_last))
# Should print [1989, 1992, 1997, 2001, 2006, 2011, 2018, 2022]
```

##Here's a detailed breakdown of the code:

- **def record_profit_years(recent_first, recent_last):** - This line defines a function named record_profit_years that takes two parameters, recent_first and recent_last, which are expected to be lists of years.

- **recent_first.reverse()**: Inside the function, the code reverses the order of the recent_first list using the reverse() method. This step arranges the years in chronological order, from oldest to most recent.

- **recent_last.extend(recent_first):** The code then extends the recent_last list by appending the reversed recent_first list. This combines both lists, ensuring that the years are in chronological order.

- **return recent_last:** Finally, the function returns the combined list, which is now in chronological order.

- **recent_first = [2022, 2018, 2011, 2006] and recent_last = [1989, 1992, 1997, 2001]:** These lines assign two lists of years to the recent_first and recent_last variables.

- **print(record_profit_years(recent_first, recent_last))**: The code calls the record_profit_years function, passing the recent_first and recent_last lists as parameters. It then prints the result, which should be the combined list of years in chronological order.

This code can be useful in scenarios where data needs to be organized chronologically, such as financial data, historical records, or any other application where time-based ordering is crucial.
